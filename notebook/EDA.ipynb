```python
df['total score']=df['math_score']+df['reading_score']+df['writing_score']
df['average']=df['total score']/3
df.head()
```

```python
df.tail()
```

```python
reading_full = df[df['reading_score'] == 100]['average'].count()
writing_full = df[df['writing_score'] == 100]['average'].count()
math_full = df[df['math_score'] == 100]['average'].count()

print(" Full marks in Maths:", math_full)
print( "Full marks in writing Writing:", writing_full)
print( "Full marks in Reading:", reading_full)
```

```python
reading_lessthan_20 = df[df['reading_score'] <= 20]['average'].count()
writing_lessthan_20= df[df['writing_score'] <= 20]['average'].count()
math_lessthan_20 = df[df['math_score'] <= 20]['average'].count()

print("  marks lessthan 20 in Maths:", math_lessthan_20)
print( "  marks lessthan 20 in writing Writing:", writing_lessthan_20)
print( "  marks lessthan 20 in Reading:", reading_lessthan_20)
```
